```python
import pandas as pd
vectors = pd.read_csv('vectors.csv')
columns = ["Original_Spock", "Original_Kirk", "Janeway", "Archer", "Sisko", "Picard", "Data", "Animated_Spock", "Animated_Kirk"]
t_vectors = vectors.drop(columns ='Unnamed: 0').transpose()
t_vectors.columns = columns
spock1_v = t_vectors['Original_Spock']['vector']
kirk1_v = t_vectors['Original_Kirk']['vector']
janeway_v = t_vectors['Janeway']['vector']
archer_v = t_vectors['Archer']['vector']
sisko_v = t_vectors['Sisko']['vector']
picard_v = t_vectors['Picard']['vector']
data_v = t_vectors['Data']['vector']
spock2_v = t_vectors['Animated_Spock']['vector']
kirk2_v = t_vectors['Animated_Kirk']['vector']
#print(x.split())
counter = 0
spock1_dict = {}
kirk1_dict = {}
janeway_dict = {}
archer_dict = {}
sisko_dict = {}
picard_dict = {}
data_dict = {}
spock2_dict = {}
kirk2_dict = {}
for i in spock1_v.split():
    vector_num = [] 
    if i[0] == '[':
        vector_num.append(float(i[1:len(i)]))
    if i[-1] == ']':
        vector_num.append(float(i[0:len(i) - 1]))
    if i[0] != '[' and i[-1] != ']':
        vector_num.append(float(i))
    for vec in vector_num:
        spock1_dict[counter] = vec
        counter = counter + 1
print(spock1_dict)
```

```python
vector_df = pd.DataFrame(vectors_dict)
vector_df_t = vector_df.transpose()
#vector_df_t['character'] = [0, 1, 2, 3, 4, 5, 0, 0, 1]
vector_df_t
chars = vector_df_t.index
```

```python
vector_df_t.to_csv('vector_df.csv')
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
data_pca = PCA().fit(vector_df_t)
plt.plot(data_pca.explained_variance_ratio_.cumsum())
plt.xlabel('Component #')
plt.ylabel('Cumulative ratio variance')
plt.title('Cumulative variance ratio explained by PCA components for phishing website statistics');
#I think it's notable below that this doesn't capture a large percentage of the variance within a short span
```

```python
data_pca_x = data_pca.transform(vector_df_t)
data_pca_x.shape
```

```python
x = data_pca_x[:, 0]
y = data_pca_x[:, 1]
pc_var = 100 * data_pca.explained_variance_ratio_.cumsum()[1]
plt.subplots(figsize=(10,8))
plt.scatter(x=x, y=y)
plt.xlabel('First component')
plt.ylabel('Second component')
plt.title(f'Phishing website summary PCA, {pc_var:.1f}% variance explained')
for s, x, y in zip(chars, x, y):
    plt.annotate(s, (x, y))
```
