```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
vectors = pd.read_csv('vector_df.csv')
vectors.drop(columns='Unnamed: 0', inplace=True)
vectors['character'] = ["Original_Spock", "Original_Kirk", "Janeway", "Archer", "Sisko", "Picard", "Data", "Animated_Spock", "Animated_Kirk"]
vectors.set_index('character', inplace=True)
index = vectors.index
```

```python
sns.set_palette("Accent")
data_pca = PCA().fit(vectors)
plt.plot(data_pca.explained_variance_ratio_.cumsum())
plt.xlabel('Component #')
plt.ylabel('Cumulative ratio variance')
plt.title('Cumulative variance ratio explained by PCA components for Star Trek dialogue vector');
```

```python
data_pca_x = data_pca.transform(vectors)
data_pca_x.shape
```

```python
x = data_pca_x[:, 0]
y = data_pca_x[:, 1]
pc_var = 100 * data_pca.explained_variance_ratio_.cumsum()[1]
plt.subplots(figsize=(10,8))
plt.scatter(x=x, y=y)
plt.xlabel('First component')
plt.ylabel('Second component')
plt.title(f'Star Trek Dialogue Vector PCA, {pc_var:.1f}% variance explained')
for s, x, y in zip(index, x, y):
    plt.annotate(s, (x, y))
```

```python
#Finally, above, you can see a PCA visualization of 86.4% of the variance captured in each character's dialogue vector.
#After everything above, the visualization makes a lot of sense.
#Picard sits right in the middle. This tallies with what we saw in the heatmap above, where he had no stark
#dissimilarities from any other character. Archer and Sisko are very close to each other and Janeway bridges the 
#gap between them and Picard. 
#It is notable in this visualization that both Kirks and both Spocks seem very differentiated from the others. It's
#also clearer in this visualization that Animated Kirk is more dissimilar from the other captains than was Original
#Kirk. Original Kirk and Animated Kirk are slightly further apart from each other than are Original Spock and 
#Animated Spock. Animated Kirk, though not moving towards Original Spock in a straight line, did move closer to 
#Original Spock relative to Original Kirk, while Animated Spock moved away from Original Kirk, relative to the 
#position of original Spock. Data, meanwhile, is quite removed from all the captains, but is closest, relatively, 
#to the Spocks and Picard.
```
